## The Prediction Logic Task

---

### Scenario

We are building a simple AI to predict the **price of a laptop** based on its **RAM (GB)**.

| RAM (GB) | Price (USD) |
|----------|-------------|
| 4        | 400         |
| 8        | 800         |
| 12       | 1200        |
| 16       | ?           |

---

### Identifying the Pattern

Looking at the table, the relationship is clear and consistent:

- 4 GB  → $400   (400 ÷ 4 = **100**)
- 8 GB  → $800   (800 ÷ 8 = **100**)
- 12 GB → $1200  (1200 ÷ 12 = **100**)

**Rule discovered:**

> **Price = RAM × 100**

This is exactly what **Linear Regression** does — it finds a simple mathematical relationship (a straight line passing through the origin) between an input variable and an output variable to make predictions.

In standard Linear Regression notation:

$$\hat{y} = m \cdot x + b$$

Here, **m = 100** (slope) and **b = 0** (intercept), where *x* is RAM and *ŷ* is the predicted price.

---

### Python Prediction Script

```python
# Step 1: Create a variable for the RAM size we want to predict
ram_size = 16

# Step 2: Apply the discovered rule — Price = RAM x 100
price_per_gb = 100

# Step 3: Calculate the predicted price
predicted_price = ram_size * price_per_gb

# Step 4: Display the result in a friendly message
print(f"The predicted price for a {ram_size}GB laptop is ${predicted_price}.")
```

---

### How This Relates to Linear Regression

| Concept | This Task |
|---|---|
| **Input feature (x)** | RAM in GB |
| **Output / label (y)** | Laptop price in USD |
| **Slope (m)** | 100 (price increase per 1 GB of RAM) |
| **Intercept (b)** | 0 (a laptop with 0 GB costs $0) |
| **Model equation** | Price = 100 × RAM |
| **Prediction** | 16 GB → **$1600** |

In a real Machine Learning workflow, a Linear Regression algorithm would **learn** this multiplier (100) automatically from the training data — we identified it manually here because the pattern is perfectly linear.
