```python
# Step 1: Create the students dictionary
students = {
    "S001": {
        "name": "Ali Hassan",
        "marks": [85, 92, 78, 90, 88]
    },
    "S002": {
        "name": "Sara Khan",
        "marks": [60, 55, 72, 65, 58]
    },
    "S003": {
        "name": "Umar Farooq",
        "marks": [95, 98, 92, 97, 99]
    }
}

print("Dictionary created successfully!")
print(students)
```

```python
# Step 2: Loop + if-elif-else + f-string Report
print("=" * 45)
print("        STUDENT PERFORMANCE REPORT")
print("=" * 45)

for student_id, info in students.items():
    name = info["name"]
    marks = info["marks"]

    # Calculate average
    average = sum(marks) / len(marks)

    # if-elif-else to categorize grade
    if average >= 90:
        category = "Exceptional"
    elif average >= 70:
        category = "Standard"
    else:
        category = "Needs Improvement"

    # Formatted report using f-strings
    print(f"\nStudent ID : {student_id}")
    print(f"Name       : {name}")
    print(f"Marks      : {marks}")
    print(f"Average    : {average:.2f}")
    print(f"Category   : {category}")

print("\n" + "=" * 45)
print("           END OF REPORT")
print("=" * 45)
```
