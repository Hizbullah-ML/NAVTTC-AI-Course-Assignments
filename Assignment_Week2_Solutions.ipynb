## Task 1: Mathematical Utilities for Data Preprocessing

This task defines three utility functions commonly used in data preprocessing pipelines.  
Each function includes a descriptive **docstring** and a **return statement** as required.

```python
import math


def calculate_circle_area(radius):
    """
    Calculate the area of a circle given its radius.

    The area is computed using the standard formula: A = π * r²

    Parameters:
        radius (float): The radius of the circle. Must be a non-negative number.

    Returns:
        float: The area of the circle.

    Example:
        >>> calculate_circle_area(5)
        78.53981633974483
    """
    return math.pi * radius ** 2


def celsius_to_fahrenheit(celsius):
    """
    Convert a temperature from Celsius to Fahrenheit.

    Uses the standard conversion formula: F = (C × 1.8) + 32

    Parameters:
        celsius (float): Temperature in degrees Celsius.

    Returns:
        float: The equivalent temperature in degrees Fahrenheit.

    Example:
        >>> celsius_to_fahrenheit(100)
        212.0
    """
    return (celsius * 1.8) + 32


def is_prime(n):
    """
    Determine whether a given integer is a prime number.

    A prime number is a natural number greater than 1 that has no positive
    divisors other than 1 and itself. This function checks divisibility
    up to the square root of n for efficiency.

    Parameters:
        n (int): The integer to check.

    Returns:
        bool: True if n is prime, False otherwise.

    Example:
        >>> is_prime(7)
        True
        >>> is_prime(9)
        False
    """
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True


# --- Test the functions ---

# Task 1.1: Circle Area
radius = 5
print(f"Circle area (radius={radius}): {calculate_circle_area(radius):.4f}")

# Task 1.2: Celsius to Fahrenheit
temp_c = 37
print(f"{temp_c}°C in Fahrenheit: {celsius_to_fahrenheit(temp_c)}°F")

# Task 1.3: Prime Check
for num in [2, 7, 10, 13, 25]:
    print(f"is_prime({num}): {is_prime(num)}")
```

## Task 2: BankAccount Class with Encapsulation

This task implements a `BankAccount` class that protects the account balance using a **private attribute** (`__balance`).  
The `withdraw` method uses an **if-else statement** to ensure sufficient funds before any deduction.

```python
class BankAccount:
    """
    A class representing a bank account with basic deposit and withdrawal operations.

    The account balance is stored as a private attribute (__balance) to prevent
    direct external modification, simulating encapsulation in Python.

    Attributes:
        __balance (float): The current account balance (private).

    Methods:
        deposit(amount)  -- Adds funds to the account.
        withdraw(amount) -- Deducts funds if balance is sufficient.
        get_balance()    -- Returns the current balance (read-only access).
    """

    def __init__(self, initial_balance=0.0):
        """
        Initialize the BankAccount with an optional starting balance.

        Parameters:
            initial_balance (float): Starting balance. Defaults to 0.0.
        """
        self.__balance = initial_balance

    def deposit(self, amount):
        """
        Deposit a positive amount into the account.

        Parameters:
            amount (float): The amount to deposit. Must be greater than zero.

        Returns:
            None
        """
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount:.2f} | New Balance: ${self.__balance:.2f}")
        else:
            print("Deposit amount must be greater than zero.")

    def withdraw(self, amount):
        """
        Withdraw a specified amount from the account if sufficient funds exist.

        Uses an if-else statement to check whether the requested withdrawal
        amount is available in the current balance before deducting it.

        Parameters:
            amount (float): The amount to withdraw. Must be greater than zero.

        Returns:
            None
        """
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn: ${amount:.2f} | Remaining Balance: ${self.__balance:.2f}")
        else:
            print(f"Insufficient funds. Attempted: ${amount:.2f} | Available: ${self.__balance:.2f}")

    def get_balance(self):
        """
        Retrieve the current account balance.

        Returns:
            float: The current balance.
        """
        return self.__balance


# --- Test the BankAccount class ---

account = BankAccount(initial_balance=500.0)
print(f"Opening Balance: ${account.get_balance():.2f}\n")

account.deposit(200.0)          # Valid deposit
account.withdraw(100.0)         # Valid withdrawal
account.withdraw(700.0)         # Should trigger insufficient funds message
account.deposit(-50)            # Invalid deposit

print(f"\nFinal Balance: ${account.get_balance():.2f}")
```
